```
df = pd.read_csv("/content/drive/My Drive/Unit 2-Climate Change/cholorophyll_temp_38N_68W.csv")
```

```
df.head()
```

```
temperature_trend = df["Temperature (C)"]
chlorophyll_trend = df["Chlorophyll (mg/m3)"]
years = df["Year"]
```

```
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
ax.scatter(years, chlorophyll_trend, color="red", s = 10)
# set x-axis label
ax.set_xlabel("Years",fontsize=10)
# set y-axis label
ax.set_ylabel("Chlorophyll (mg/m3)",color="red",fontsize=14)
# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
plt.xticks(range(1998,2017, 2))
# make a plot with different y-axis using second axis object
ax2.scatter(years, temperature_trend,color="blue", s = 10)
ax2.set_ylabel("Temperature (C)",color="blue",fontsize=14)
plt.title("Chlorophyll and Temperature Trends from 1995-2017", fontsize=14)
plt.show()

```

```
plt.scatter(temperature_trend, chlorophyll_trend, color = "orange", s = 25)
plt.title("Chrolophyll Production from 1995-2017")
plt.xlabel("Temperature (C)")
plt.ylabel("Chlorophyll (mg/m3)")
plt.xlim(10,25)
plt.xticks(range(10,25,1))
plt.show()
```

```
plt.scatter(temperature_2002, chlorophyll_2002)
plt.title("Chrolophyll Production in 2002")
plt.xlabel("Temperature (C)")
plt.ylabel("Chlorophyll (mg/m3)")
plt.xscale("linear")
plt.show()
```

```
plt.scatter(temperature_2017, chlorophyll_2017)
plt.title("Chrolophyll Production in 2017")
plt.xlabel("Temperature (C)")
plt.ylabel("Chlorophyll (mg/m3)")
plt.xscale("linear")
plt.show
```
